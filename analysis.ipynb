#### Test 1: Whether the average response time across complaint types is similar or not.

```python
# Importing required function for ANOVA test
from scipy.stats import f_oneway

# Grouping the closing times by complaint type â€” each group becomes a list of values
grouped_times = df.groupby('Complaint Type')['Request_Closing_Time'].apply(list)

# Performing a one-way ANOVA test on all the grouped lists
f_stat, p_value = f_oneway(*grouped_times)

# Printing F-statistic and P-value to interpret test results
print("F-statistic:", f_stat)
print("P-value:", p_value)

# Interpretation based on significance level (alpha = 0.05)
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in closing times among complaint types.")
else:
    print("Fail to reject the null hypothesis: No significant difference in closing times among complaint types.")
```

#### Test 2: Whether the type of complaint or service requested, and location are related.

```python
# Importing the chi2_contingency function for the Chi-Square Test
from scipy.stats import chi2_contingency

# Creating a contingency table: rows = Complaint Type, columns = Borough, values = count of records
contingency_table = pd.crosstab(df['Complaint Type'], df['Borough'])

# Performing the Chi-square test on the contingency table
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Outputting the test results
print("Chi-square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)

# Interpreting the p-value using 0.05 as the significance level (alpha)
if p_value < 0.05:
    print("Reject the null hypothesis: Complaint type and borough are dependent (associated).")
else:
    print("Fail to reject the null hypothesis: Complaint type and borough are independent (not associated).")
```
